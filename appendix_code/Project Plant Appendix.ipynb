```python
def L_system(N, V='', C='', A='', R=['']):
    '''
    This function uses a given set of variables, constants, rules and an axiom to iterate lines which builds up on the previous one.
    
    input:
        N:    number of times to run the function. e.g. 7
        V:          string which contains all the variables. e.g. 'AB' for variables 'A' and 'B'
        C:          string which contains all the constants. e.g. '+-' for constants '+' and '-'
        A:              string which contains the initial state. e.g. 'A+B-AA'.
                                Can only contain variables and constants characters.
        R:              list of strings which contains the rules for each constants in the order the constants
                                are given, e.g. ['AA-', 'AB+-A'] for rule 'A' --> 'AA-' and rule 'B' --> 'AB+-A'.
                                Note that there need to be as many rules as variables.
                                Can only contain variables and constants characters.
    output:
        List of N+1 strings, one string for each iteration and the original axiom string.
    '''
    variables = list(V) #separate the variables into list of one character strings
    constants = list(C) #separate the constants into list of one character strings
    axiom = list(A) #separate the axiom into list of one character strings
    rules = [] #separate rules into list of list of one character strings
    for ll in R:
        rules.append(list(ll))
    saved = []  #save the axiom as the first line to return
    line_before = axiom
    saved.append(''.join(axiom))
    for jj in range(N):   #iterate over the rules for as many interation as wanted
        line_after = []
        for ii in range(len(line_before)):  #go over each characters in the string to apply rules to each characters
            for pp in range(len(variables)):    #apply rules to each variables
                if line_before[ii] == variables[pp]:
                    line_after.extend(rules[pp])
            for kk in constants:    #place each constants in the new string list. Constants do not change.
                if line_before[ii] == kk:
                    line_after.extend(kk)
        saved.append(''.join(line_after))   #save the new list of one character strings as a single string
        line_before = line_after    #set the input string for the next iteration as the previous string
    return saved

#Draw with vector drawings using Turtle Graphics (common for L-System)
#draw binary fractal tree:
def draw_tree(Number_of_iteration, length=10, angle=45):
    tree = L_system(Number_of_iteration, V='01', C='[]', A='0', R=['1[0]0', '11'])
    #take canvas size to be 250 by 250 units
    pen = turtle.Turtle()
    pen.speed(0) #control speed at which pen is moving
    pen.color('black') #controls color of pen
    pen.hideturtle() #hide the cursor on the canvas

    tree_draw = list(tree[-1])

    pen.setheading(90) #start by drawing up
    pen.penup() #pen does not draw on canvas in that state, but does move
    pen.setposition(0, -250) #start from bottom centre of page
    pen.pendown() #pen draws on canvas in that state
    position_save = []
    angle_save = []
    for ii in range(len(tree_draw)):
        if tree_draw[ii] == '0':
            pen.pendown()
            pen.forward(length)
            pen.penup()
        elif tree_draw[ii] == '1':
            pen.pendown()
            pen.forward(length)
            pen.penup()
        elif tree_draw[ii] == '[':
            pen.left(angle)
            position_save.append(pen.position())
            angle_save.append(pen.heading())
        elif tree_draw[ii] == ']':
            pen.setposition(position_save[-1])
            pen.setheading(angle_save[-1])
            pen.right(2*angle)
            position_save.pop() #remove last saved position once used
            angle_save.pop()

    turtle.exitonclick() #keep the drawing open until clicking once done
    return

#draw koch curve:
def draw_koch(Number_of_iteration, length=10, angle=90):
    koch = L_system(Number_of_iteration, V='F', C='+-', A='F', R=['F+F-F-F+F'])
    #take canvas size to be 250 by 250 units
    pen = turtle.Turtle()
    pen.speed(0) #control speed at which pen is moving
    pen.color('black') #controls color of pen
    pen.hideturtle() #hide the cursor on the canvas

    koch_draw = list(koch[-1])

    pen.setheading(0) #start by drawing right
    pen.penup() #pen does not draw on canvas in that state, but does move
    pen.setposition(-250, -250) #start from bottom left of page
    pen.pendown() #pen draws on canvas in that state

    for ii in range(len(koch_draw)):
        if koch_draw[ii] == 'F':
            pen.pendown()
            pen.forward(length)
            pen.penup()
        elif koch_draw[ii] == '+':
            pen.left(angle)
        elif koch_draw[ii] == '-':
            pen.right(angle)

    turtle.exitonclick() #keep the drawing open until clicking once done
    return

#draw triangle
def draw_triangle(Number_of_iteration, length=10, angle=120):
    triangle = L_system(Number_of_iteration, V='FG', C='+-', A='F-G-G', R=['F-G+F+G-F', 'GG'])
    #take canvas size to be 250 by 250 units
    pen = turtle.Turtle()
    pen.speed(0) #control speed at which pen is moving
    pen.color('black') #controls color of pen
    pen.hideturtle() #hide the cursor on the canvas

    triangle_draw = list(triangle[-1])

    pen.setheading(0) #start by drawing right
    pen.penup() #pen does not draw on canvas in that state, but does move
    pen.setposition(-250, 250) #start from top left of page
    pen.pendown() #pen draws on canvas in that state

    for ii in range(len(triangle_draw)):
        if triangle_draw[ii] == 'F':
            pen.pendown()
            pen.forward(length)
            pen.penup()
        elif triangle_draw[ii] == 'G':
            pen.pendown()
            pen.forward(length)
            pen.penup()
        elif triangle_draw[ii] == '+':
            pen.left(angle)
        elif triangle_draw[ii] == '-':
            pen.right(angle)

    turtle.exitonclick() #keep the drawing open until clicking once done
    return

#draw dragon curve
def draw_dragon(Number_of_iteration, length=10, angle=90):
    dragon = L_system(Number_of_iteration, V='FG', C='+-', A='F', R=['F+G', 'F-G'])
    
    pen = turtle.Turtle()
    pen.speed(0) #control speed at which pen is moving
    pen.color('black') #controls color of pen
    pen.hideturtle() #hide the cursor on the canvas

    dragon_draw = list(dragon[-1])

    pen.setheading(0) #start by drawing right
    pen.penup() #pen does not draw on canvas in that state, but does move
    pen.setposition(0, 0) #start from top left of page
    pen.pendown() #pen draws on canvas in that state

    for ii in range(len(dragon_draw)):
        if dragon_draw[ii] == 'F':
            pen.pendown()
            pen.forward(length*2/3)
            pen.penup()
        elif dragon_draw[ii] == 'G':
            pen.pendown()
            pen.forward(length*2/3)
            pen.penup()
        elif dragon_draw[ii] == '+':
            pen.left(angle/2)
            pen.pendown()
            pen.forward(length*np.sqrt(2)/3)
            pen.penup()
            pen.left(angle/2)
        elif dragon_draw[ii] == '-':
            pen.right(angle/2)
            pen.pendown()
            pen.forward(length*np.sqrt(2)/3)
            pen.penup()
            pen.right(angle/2)

    turtle.exitonclick() #keep the drawing open until clicking once done
    return

#draw plant
def draw_fractal_plant(Number_of_iteration, length=10, angle=25):
    fractal_plant = L_system(Number_of_iteration, V='XF', C='+-[]', A='X', R=['F+[[X]-X]-F[-FX]+X', 'FF'])
    #take canvas size to be 250 by 250 units
    pen = turtle.Turtle()
    pen.speed(0) #control speed at which pen is moving
    pen.color('black') #controls color of pen
    pen.hideturtle() #hide the cursor on the canvas

    fractal_plant_draw = list(fractal_plant[-1])

    pen.setheading(90) #start by drawing up
    pen.penup() #pen does not draw on canvas in that state, but does move
    pen.setposition(-150, -250) #start from bottom centre of page
    pen.pendown() #pen draws on canvas in that state
    position_save = []
    angle_save = []
    for ii in range(len(fractal_plant_draw)):
        if fractal_plant_draw[ii] == 'X':
            pen.pendown()
            pen.penup()
        elif fractal_plant_draw[ii] == 'F':
            pen.pendown()
            pen.forward(length)
            pen.penup()
        elif fractal_plant_draw[ii] == '[':
            position_save.append(pen.position())
            angle_save.append(pen.heading())
        elif fractal_plant_draw[ii] == ']':
            pen.setposition(position_save[-1])
            pen.setheading(angle_save[-1])
            position_save.pop() #remove last saved position once used
            angle_save.pop() #remove last saved angle once used
        elif fractal_plant_draw[ii] == '+':
            pen.left(angle)
        elif fractal_plant_draw[ii] == '-':
            pen.right(angle)

    turtle.exitonclick() #keep the drawing open until clicking once done
    return

#draw_triangle(4, length=20)
#draw_tree(5)
#draw_koch(3)
#draw_dragon(10, length=5)
#draw_fractal_plant(3, length=10)
```
